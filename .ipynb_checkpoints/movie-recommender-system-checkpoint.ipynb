```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.shape
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
credits.shape
```

```python
moviez = movies.merge(credits,on='title')
moviez.shape
```

```python
moviez.columns
```

```python
#Important columns
#genre
#id
#keywords
#title
#overview
#cast
#crew
moviez=moviez[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
moviez.head()
```

```python
moviez.isnull().sum()
```

```python
moviez.dropna(inplace=True)
```

```python
moviez.duplicated().sum()
```

```python
moviez.iloc[0].genres
```

```python
import ast   #ast.literal_eval is used to convert a string to a list
```

```python
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i["name"])
    return list
```

```python
moviez['genres']=moviez['genres'].apply(convert)
moviez.head()
```

```python
moviez['keywords']=moviez['keywords'].apply(convert)
moviez.head()
```

```python
def convert3(obj):
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list
```

```python
moviez['cast']=moviez['cast'].apply(convert3)
```

```python
moviez.head()
```

```python
def get_director(obj):
    list=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i["name"])
            break
    return list
```

```python
moviez['crew'] = moviez['crew'].apply(get_director)
moviez.head()
```

```python
moviez['overview'] = moviez['overview'].apply(lambda x:x.split())
```

```python
moviez.head()
```

```python
moviez['genres']=moviez['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
moviez['keywords']=moviez['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
moviez['cast']=moviez['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
moviez['crew']=moviez['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
moviez.head()
```

```python
moviez['tags'] = moviez['overview'] + moviez['genres'] + moviez['keywords'] + moviez['cast'] + moviez['crew']
moviez.head()
```

```python
new_df = moviez[["movie_id", "title", "tags"]]
new_df.head()
```

```python
# new_df['tags']=
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemmer(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
print(stemmer("'In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron'"))
```

```python
new_df['tags']=new_df['tags'].apply(stemmer)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
np.set_printoptions(threshold=np.inf)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    similarity_of_movies = similarity[movie_index]
    sorted_movies = sorted(list(enumerate(similarity_of_movies)), reverse=True, key=lambda x:x[1])[1:6]
    for i in sorted_movies:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
